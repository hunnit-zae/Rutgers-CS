```python
import csv
file = open("pokemonTrain.csv","r")
reader=csv.reader(file)
pokemon=[]
# read pokemon database
for line in reader:
    pokemon.append(line)
file.close()
#print (pokemon)
# 1.1 finding the percentage of fire type pokemons over 40
Totalfiretype=0
firetypeover40=0
for row in pokemon:
    if row [4]=="fire":
        Totalfiretype+=1
        if float (row [2])>=40:
            firetypeover40+=1
fireTypeRatio = round (firetypeover40/Totalfiretype*100)
with open("pokemon1.txt", "w") as f:
    f.write("Percentage of fire type Pokemons at or above level 40 = "+ str(fireTypeRatio))
# 1.2, 1.3 filling in the missing weaknesses, values
weakness=dict()
for row in pokemon[1:]:
    w = row [5]
    t = row [4]
    if w not in weakness:
        weakness [w]=[]
    if t != "NaN":
        weakness [w].append(t)
#print (weakness)

from collections import Counter
 
def findType(lst: list) -> str:
    typeCnt = Counter(lst)
    type = ""
    cnt = 0
    for key, val in typeCnt.items():
        if val > cnt or (val == cnt and key < type):
               type = key
        
    return type

for key in weakness:
    weakness[key]=findType(weakness[key])
#print(weakness)
for row in pokemon:
    if row [4]=="NaN":
        row [4]=weakness[row[5]]

import math
attackover40=[]
defenseover40=[]
healthover40=[]
attackunder=[]
defenseunder=[]
healthunder40=[]
totalhp=0
stage3=0
for row in pokemon[1:]:
    level=float (row[2])
    attack=float (row[6])
    defense=float (row[7])
    health=float (row[8])
    stage=float (row[9])
    if level >40:
        if not math.isnan(attack):
            attackover40.append(attack)
        if not math.isnan(defense):
            defenseover40.append(defense)
        if not math.isnan(health):
            healthover40.append(health)
    else:
        if not math.isnan(attack):
            attackunder.append(attack)
        if not math.isnan(defense):
            defenseunder.append(defense)
        if not math.isnan(health):
            healthunder40.append(health)
    if stage==3:
        totalhp+=health
        stage3+=1
averageattack40=round(sum (attackover40)/len (attackover40),1)
averagedefense40=round(sum (defenseover40)/len (defenseover40),1)
averagehealth40=round(sum (healthover40)/len (healthover40),1)
averageattackunder=round(sum (attackunder)/len (attackunder),1)
averagedefense=round(sum (defenseunder)/len (defenseunder),1)
averagehealth=round(sum (healthunder40)/len (healthunder40),1)
#print (averageattack40,averagedefense40,averagehealth40)
#print (stage3)
for row in pokemon[1:]:
    level=float (row[2])
    if row[6]=="NaN":
        if level>40:
            row[6]=averageattack40
        else: 
            row[6]=averageattackunder
    if row[7]=="NaN":
        if level>40:
            row[7]=averagedefense40
        else:
            row[7]=averagedefense
    if row[8]=="NaN":
        if level>40:
            row[8]=averagehealth40
        else:
            row[8]=averagehealth
file=open("pokemonResult.csv","w")
writer=csv.writer(file)
writer.writerows(pokemon)
file.close()

# 1.4 personalities
from collections import defaultdict
personality=defaultdict(list)
for row in pokemon[1:]:
    type=row[4]
    p=row[3]
    if p not in personality[type]:
        personality[type].append(p)
for t in personality:
    personality[t].sort()
types=sorted(list(personality))
#print(types)
with open("pokemon4.txt", "w") as f:
    f.write("Pokemon type to personality mapping:")
    f.write("\n\n")
    for t in types:   
        p_list = ", ".join(personality[t])
        f.write(f"   {t}: {p_list}\n")

# 1.5 average hp
totalHP = 0.0
stage3 = 0
for row in pokemon[1:]:
    if float(row[9]) == 3.0:
        totalHP += float(row[8])
        stage3 = stage3 + 1
avgHP = round(totalHP / stage3)
with open("pokemon5.txt", "w") as f:
    f.write("Average hit point for Pokemons of stage 3.0 = " + str(avgHP))
```
