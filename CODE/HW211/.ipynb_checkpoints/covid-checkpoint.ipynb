```python
import csv
file = open("covidTrain.csv","r")
reader=csv.reader(file)
covidTrain=[]
for line in reader:
    covidTrain.append(line)
file.close()
import math
latitude=dict()
longitude=dict()
for row in covidTrain [1:]:
    age=row[1]
    if '-' in str (age):
        i = age.find('-')
        start = int(age[:i])
        end = int(age[i+1:])
        avg = (start+end)/2
        avg = round(avg)
        row[1]=avg
    #print(row[1])
    for j in range(8,11):
        date = row[j]
        month = date[3:5]
        day = date[0:2]
        year = date[6:]
        row[j]=month+'.'+day+'.'+year
        #print (row [j])
    province=row[4]
    lat=float (row[6])
    lon=float (row[7])
    if  not math.isnan(lat) and province in latitude:
        latitude[province] = (latitude [province][0]+lat, latitude [province][1]+1)
    if not math.isnan(lon) and province in longitude:
        longitude[province] = (longitude[province][0]+lon, longitude[province][1]+1)
    if province not in latitude:
        latitude[province] = (0,0)
    if province not in longitude:
        longitude[province] = (0,0)
#print(latitude)
#print(longitude)
for p in latitude.keys():
    latitude[p]=round(latitude[p][0]/latitude[p][1],2)
for p in longitude.keys():
    longitude[p]=round(longitude[p][0]/longitude[p][1],2)  
#print(latitude)
#print(longitude)
provinces=dict()
symptoms=dict()
for row in covidTrain [1:]:
    sym=row[11]
    province=row[4]
    lat=float (row[6])
    lon=float (row[7])
    city=row[3]
    if sym != 'NaN':
        s=sym.split(';')
        for item in s:
            item = item.strip()
            if province in symptoms:
                if item in symptoms[province]:
                    symptoms [province][item]+=1
                else:
                    symptoms [province][item]=1
            else:
                symptoms[province]=dict()
    if city != 'NaN':
        if province in provinces:
            if city in provinces[province]:
                provinces [province][city]+=1
            else:
                provinces [province][city]=1
        else:
            provinces [province]=dict()
    if math.isnan(lat):
        row[6]=latitude[province]
    if math.isnan(lon):
        row[7]=longitude[province]
#print(provinces)
for p in provinces:
    x = provinces[p]
    provinces[p] ={k: v for k, v in sorted(x.items(), key=lambda item: item[1])}
#print(symptoms)
for row in covidTrain [1:]:
    province=row[4]
    city=row[3]
    m = -1
    name = ''
    if city == 'NaN':
        x = provinces[province]
        for c in x:
            if x[c] > m:                
                m = x[c]
                name = c
            elif x[c] == m and c < name:
                m = x[c]
                name = c
        row[3] = name
    #print(row[3])
for row in covidTrain [1:]:
    province=row[4]
    sym=row[11]
    m = -1
    name = ''
    if sym == 'NaN':
        x = symptoms[province]
        for c in x:
            if x[c] > m:                
                m = x[c]
                name = c
            elif x[c] == m and c < name:
                m = x[c]
                name = c
        row[11] = name
    #print(row[11])
for row in covidTrain [1:]:
    row[6] = round(float(row[6]),2)
    row[7] = round(float(row[7]),2)
with open('covidResult.csv','w') as f:
    for row in covidTrain:
        s = ''
        for i in row:
            s+=str(i)+','
        f.write(s)
        f.write('\n')
f.close()
```
