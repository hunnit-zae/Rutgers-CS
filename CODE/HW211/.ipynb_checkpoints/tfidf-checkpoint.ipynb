```python
def stop(f):
    words=[]
    with open(f,'r') as file:
        for line in file:
            for w in line.split():
                words.append(w)
    file.close()
    return words
def clean(f):
    stopwords=stop('stopwords.txt')
    words=[]
    with open(f,'r') as file:
        for line in file:
            line = line.split()
            for w in line:
                w = w.lower()
                if w in stopwords or 'http://' in w or 'https://' in w or w=='':
                    continue
                else:
                    w=w.lower()
                    if w [-3:]=='ing':
                        w = w[:-3]
                    if w [-2:]=='ly':
                        w = w[:-2]
                    if w [-4:]=='ment':
                        w = w[:-4]
                    if w [-2] == '\'':
                        w = w[:-2] + w[-1]
                    if w [-1:] == ',':
                        w = w[:-1]    
                    word = ''
                    for c in w:
                        if c != ' ' and c != '{' and c != '}' and c != '.' and c != ',' and c != '(' and c != ')' and c != '*' and c != '-' and c != '\"':
                            word += c
                    if word not in stopwords:
                        words.append(word.lower())
    file.close()
    s = 'preproc_'+f
    with open(s,'w')as file:
        for w in words:
            file.write(w)
            file.write(' ')
def tf(f):
    words=dict()
    total = 0
    with open(f,'r')as file:        
        for line in file:
            line=line.split()
            for w in line:
                if w in words:
                    words[w]+=1
                else:
                    words[w]=1
                total+=1
    for w in words.keys():
        words [w]/=total
    return words
import math
car = []
with open ('tfidf_docs.txt','r')as file:
    for line in file:
        line=line.split()
        #print(line)
        clean(line[0])
        s='preproc_'+line[0]
        fin = 'tfidf_'+line[0]
        car.append(tf(s))
    #print(car)

        idf=dict()
        for doc in car:
            for w in doc:
                if w not in idf:
                    count=0
                    for doc in car:
                        if w in doc:
                            count+=1
                    score=math.log(len(car)/count)
                    idf [w]=round(score,2)+1
        #print(idf)
        tfidf=dict()
        for doc in car:
            #print(doc)
            for w in doc:
                if w not in tfidf:
                    tfidf[w]=doc[w]*idf[w]
        #print(tfidf)
        tfidf = {k: v for k, v in sorted(tfidf.items(), key=lambda item: item[1], reverse = True)}
        i = 0
        with open (fin, 'w') as file:
            for k in tfidf.keys():
                if i == 5:
                    break
                file.write(k)
                file.write(' ')
                file.write(str(round(tfidf[k], 2)))
                file.write(' ')
                i += 1
```
