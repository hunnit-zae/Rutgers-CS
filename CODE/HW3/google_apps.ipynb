```python
dataframe1 = dataframe1[~dataframe1.isin(['Varies with device']).any(axis=1)]
print(dataframe1)
```

```python
#PREPROCESS 3 -- WARNING SLOW RUNTIME
#count=0
for index,row in dataframe1.iterrows():
    if (re.findall(r'\d+\.\d',str(row))):
        #print(dataframe1.loc[])
        row = re.findall(r'\d+\.\d',str(row))[0]
        #print(dataframe1.loc(index, 'Android Ver'))
    else:
        row = str(row)
    dataframe1['Android Ver'].loc[index] = row
    #print(str(row))
    
dataframe1
```

```python
#PREPROCESS 4

for index,row in dataframe1.iterrows():
    my_string = dataframe1['Installs'].loc[index]
    #print(my_string)
    my_string = my_string.replace(",","")
    my_string = my_string.replace("+","")
    #print(my_string)
    dataframe1['Installs'].loc[index] = my_string

dataframe1
```

```python
#PREPROCESS 5

for index,row in dataframe1.iterrows():
    if(pd.isnull(dataframe1['Rating'].loc[index])):
        if((dataframe1['Reviews'].loc[index] < 100) and (int(dataframe1['Installs'].loc[index]) < 50000)):
            dataframe1 = dataframe1.drop(index)
        else:
            cat = dataframe1['Category'].loc[index]
            avg = dataframe1[dataframe1.Category == cat].mean()
            dataframe1['Rating'].loc[index] = avg

dataframe1
```

```python
#PREPROCESS 6

for index,row in dataframe1.iterrows():
    if('M' in str(dataframe1['Size'].loc[index])):
        value = str(dataframe1['Size'].loc[index]).replace('M', '')
        value2 = float(value)
        value2 = value2*1000000
        value3 = int(value2)
        dataframe1['Size'].loc[index] = value3
    if(('K' in str(dataframe1['Size'].loc[index])) or ('k' in str(dataframe1['Size'].loc[index]))):
        value = str(dataframe1['Size'].loc[index]).replace('k', '')
        value2 = float(value)
        value2 = value2*1000
        value3 = int(value2)
        dataframe1['Size'].loc[index] = value3
```

```python
#Analysis 2
dataframe2 = dataframe1[dataframe1.Type == 'Free']
dataframe2
#print(dataframe2['Ratings'].max())
def findTop3(col,dataframe):
    newDF = pd.DataFrame()
    newDF.append(dataframe[col].max())
    dataframe.drop(dataframe[col].max().index())
    newDF.append(dataframe[col].max())
    dataframe.drop(dataframe[col].max().index())
    newDF.append(dataframe[col].max())
    return newDF

findTop3('Rating',dataframe2)
findTop3('Installs',dataframe2)
findTop3('Reviews',dataframe2)
    
```
